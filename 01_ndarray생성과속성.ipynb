```python
# 배열 속성
arr1 = np.array([1, 2, 3, 4])
arr1.shape
```

```python
arr2 = np.array(
    [[1, 2, 3],
    [4, 5, 6]]
)
arr2.shape
```

```python
arr3 = np.array(
    [[[1, 2], 
     [3, 4]],
     [[5, 6],
     [7, 8]]
     ]
)
arr3.shape
```

```python
# ndim (차원을 알려주는)
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
```

```python
# size - 배열의 총 요소 수를 알려준다
arr1 = np.size([1, 2, 3, 4])
print(arr1)
```

```python
arr3_size = np.size(arr3)
print(arr3_size)
```
